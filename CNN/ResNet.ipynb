# ResNet - Image Classification

- **Reference:** https://keras.io/api/applications/resnet/

```python
import tensorflow
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
from IPython.display import Image, display
```

```python
model = ResNet50(weights='imagenet')
```

```python
def predict_image(model, image_path, top_k=3):
   """Predict image classification using pretrained model"""
   img = image.load_img(image_path, target_size=(224, 224))
   display(Image(image_path))
   x = image.img_to_array(img)
   x = np.expand_dims(x, axis=0)
   x = preprocess_input(x)
   preds = model.predict(x)
   return decode_predictions(preds, top=top_k)[0]
```

```python
img_path = 'pictures/1.jpg'
predictions = predict_image(model, img_path)
for _, name, prob in predictions:
   print(f"{name}: {prob:.2%}")
```
